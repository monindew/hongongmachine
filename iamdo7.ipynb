```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
print(iris)
print(iris.feature_names)
```

```python
import pandas as pd
df = pd.DataFrame(data=iris.data, columns = iris.feature_names)
df.info()
```

```python
iris_petal = iris.data[:,[2,3]]
iris_sepal = iris.data[:,[0,1]]
labels = iris.target
```

```python
import matplotlib.pyplot as plt
plt.scatter(iris_petal[:,0], iris_petal[:,1], c="b", alpha=0.5)
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.show()
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters = 3 )
Kmean.fit(iris_petal)
```

```python
print(Kmean.cluster_centers_)
print(Kmean.labels_)
```

```python
colours = ["tab:blue", "tab:green", "tab:orange"]
center_colours = ["b", "g", "r"]
for i in range(len(Kmean.labels_)):
  label = Kmean.labels_[i]
  plt.scatter(iris_petal[i][0],
              iris_petal[i][1],
              c = colours[label],
              alpha=0.5)

for i in range(len(Kmean.cluster_centers_)):
  plt.scatter(Kmean.cluster_centers_[i][0],
              Kmean.cluster_centers_[i][1],
              s = 200,
              c = center_colours[i])
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.show()
```

```python
inertia_a = []
k_range = range(1, 10)

for k in k_range:
  kmeans = KMeans(n_clusters = k, random_state=21)
  kmeans.fit(iris_petal)
  inertia = kmeans.inertia_
  inertia_a.append(inertia)

plt.plot(k_range, inertia_a, marker = "o")
plt.xlabel("number of clusters, k", fontsize =13)
plt.ylabel("inertia", fontsize = 13)
plt.show()
```
