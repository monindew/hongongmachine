```python
penguins_encoding = pd.get_dummies(penguins, columns=["island", "sex"])
penguins_encoding.head()
```

```python
X = penguins_encoding.drop(["species"], axis=1)
y = penguins_encoding["species"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=11)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
```

```python
knn.fit(X_train, y_train)
```

```python
print(knn.score(X_train, y_train))
```

```python
y_pred = knn.predict(X_test)
```

```python
y_pred[:20]
```

```python
print('knn accuracy:{:.3f}'.format(knn.score(X_test, y_test)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
k_range = range(1,31)
scores = []
for k in k_range:
  knn = KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train, y_train)
  scores.append(knn.score(X_test,y_test))
```

```python
plt.plot(k_range,scores)
plt.xlabel('value of K')
plt.ylabel('Accuracy Score')
plt.xticks(np.arange(0,31,5))
plt.grid()
plt.title('k-NN Accuracy score')
plt.show()
```

```python
for value, score in enumerate(scores):
  print(f'k = {value+1}일때 : accuracy{score:.3f}')
```
